## Q10: 
Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! 

```python
q10_df = pd.read_sql_query("SELECT name, SUM(booking_cost) AS revenue FROM (SELECT f.name, CASE WHEN b.memid = 0 THEN f.guestcost*b.slots ELSE f.membercost*b.slots END AS booking_cost FROM Bookings AS b INNER JOIN Facilities AS f USING (facid)) AS subq GROUP BY name HAVING revenue < 1000 ORDER BY revenue", engine)

q10_df
```

 ## Q11: 
 Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
q11_df = pd.read_sql_query("SELECT mem_name, m.surname|| ', '||m.firstname AS recommedby FROM (SELECT memid, surname|| ', '||firstname AS mem_name, recommendedby FROM Members WHERE recommendedby <> '' ORDER BY mem_name) AS subq LEFT JOIN Members AS m ON subq.recommendedby = m.memid", engine)

q11_df
```

## Q12: 
Find the facilities with their usage by member, but not guests 

```python
q12_df = pd.read_sql_query("SELECT f.name, COUNT(b.memid) AS number_of_users FROM Bookings AS b INNER JOIN Facilities as f ON b.facid = f.facid WHERE memid <> 0 GROUP BY f.name", engine)

q12_df
```

## Q13: 
Find the facilities usage by month, but not guests

```python
q13_df = pd.read_sql_query("SELECT strftime('%m', starttime) AS month, f.name, COUNT(b.memid) AS number_of_users FROM Bookings AS b INNER JOIN Facilities as f ON b.facid = f.facid WHERE memid <> 0 GROUP BY f.name, month ORDER BY month, f.name", engine)

q13_df
```
